```python
import numpy as np
import matplotlib.pyplot as plt
base_path = "data/"
acc_prefix = "nAcc_"
time_prefix = "nTime_"

datasets=["gas", "elec", "abruptInsects"]
datasets2=["covtype", "airlines"]
colors=["red", "green", "blue"]
```

```python
data_acc=[]
data_time=[]
index=[]

for dataset in datasets:
    data_acc.append(np.loadtxt(base_path+acc_prefix+dataset))
    data_time.append(np.loadtxt(base_path+time_prefix+dataset))
    index.append([1+x*5 for x in range(0,len(data_acc[-1]))])
```

```python
_, ax1 = plt.subplots(figsize=(10, 3))
ax1.set_title("")
ax2 = ax1.twinx()
ax1.plot(index[0][:100], data_acc[0][:100], color="green", marker="x", alpha=0.5, linestyle=(0, (1, 3)), markersize=4, label='Gas')
ax2.plot(index[0][:100], data_time[0][:100], color="blue", marker="x", alpha=0.5, linestyle=(0, (1, 3)), markersize=4, label='Gas')
ax1.plot(index[1][:100], data_acc[1][:100], color="green", marker="s", alpha=0.5, linestyle=(0, (5, 6)), markersize=4, label='Electricity')
ax2.plot(index[1][:100], data_time[1][:100], color="blue", marker="s", alpha=0.5, linestyle=(0, (5, 6)), markersize=4, label='Electricity')
ax1.plot(index[2][:100], data_acc[2][:100], color="green", marker="o", alpha=0.5, markersize=4, label='Abrupt Insects')
ax2.plot(index[2][:100], data_time[2][:100], color="blue", marker="o", alpha=0.5, markersize=4, label='Abrupt Insects')

ax1.set_xlabel('Retraining after n samples')
ax1.set_ylabel('Accuracy in %', color='g')
ax2.set_ylabel('Running Time in s', color='b')
ax1.legend(loc="upper center")
ax2.legend(loc="upper right")
# Adjust layout to prevent overlapping titles
plt.tight_layout()
plt.savefig('baselines.png', dpi=300)
```
